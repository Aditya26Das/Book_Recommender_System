```python
category_mapping = {'Fiction' : "Fiction",
 'Juvenile Fiction': "Children's Fiction",
 'Biography & Autobiography': "Nonfiction",
 'History': "Nonfiction",
 'Literary Criticism': "Nonfiction",
 'Philosophy': "Nonfiction",
 'Religion': "Nonfiction",
 'Comics & Graphic Novels': "Fiction",
 'Drama': "Fiction",
 'Juvenile Nonfiction': "Children's Nonfiction",
 'Science': "Nonfiction",
 'Poetry': "Fiction"}

df["simple_categories"] = df["categories"].map(category_mapping)
```

```python
import torch
from transformers import pipeline # type: ignore

fiction_categories = ["Fiction", "Nonfiction"]

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

result = classifier("This is a story about a magical kingdom and a brave hero.", candidate_labels=fiction_categories)

print(result)
```

```python
max_index = np.argmax(result["scores"]) # type: ignore
predicted_category = result["labels"][max_index] # type: ignore
```

```python
max_index, predicted_category
```

```python
def predict_category(text:str, categories:list[str]) -> str:
    """
    Predict the category of a given text using zero-shot classification.
    
    Args:
        text (str): The input text to classify.
        categories (list[str]): List of candidate categories.
        
    Returns:
        str: The predicted category.
    """
    result = classifier(text, candidate_labels=categories)
    max_index = np.argmax(result["scores"]) # type: ignore
    return result["labels"][max_index] # type: ignore
```

```python
from tqdm import tqdm
```

```python
actual_categories = []
predicted_categories = []

for i in tqdm(range(0,300)):
    text = df.loc[df["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
    
    predicted_categories.append(predict_category(text=text, categories=["Fiction", "Nonfiction"]))
    actual_categories.append("Fiction")
```

```python
for i in tqdm(range(0,300)):
    text = df.loc[df["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    
    predicted_categories.append(predict_category(text=text, categories=["Fiction", "Nonfiction"]))
    actual_categories.append("Nonfiction")
```

```python
len(predicted_categories), len(actual_categories)
```

```python
cnt = 0
for i in tqdm(range(0,len(predicted_categories))):
    if predicted_categories[i] == actual_categories[i]:
        cnt += 1
```

```python
print(f"Accuracy score of zero-shot classification model: {(cnt/600)*100}\n")
```

```python
isbns = []
predict_categories = []

missing_cats = df.loc[df["simple_categories"].isna(), ["isbn13","description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats["description"][i]
    predict_categories.append(predict_category(sequence, fiction_categories))
    isbns.append(missing_cats["isbn13"][i])
```
